```python
from datetime import datetime
import pytz

def sugerir_comida_por_pais():
    pais = input("¿En qué país estás? (Ej: España, México, Argentina, Chile): ").strip().lower()

    zonas = {
        "españa": "Europe/Madrid",
        "mexico": "America/Mexico_City",
        "méxico": "America/Mexico_City",
        "argentina": "America/Argentina/Buenos_Aires",
        "chile": "America/Santiago",
        "colombia": "America/Bogota",
        "peru": "America/Lima",
        "perú": "America/Lima",
        "usa": "America/New_York",
        "estados unidos": "America/New_York"
    }

    if pais not in zonas:
        print("No tengo registrada la zona horaria de ese país.")
        return

    zona = pytz.timezone(zonas[pais])
    hora_local = datetime.now(zona).hour

    opciones = {
        "Desayuno": ["Huevos benedictinos", "Avena con frutas", "Tostadas con aguacate"],
        "Comida": ["Pechuga de pollo al grill", "Pasta boloñesa", "Ensalada César con salmón"],
        "Merienda": ["Yogur griego con nueces", "Batido de proteínas", "Fruta de temporada"],
        "Cena": ["Tacos de pescado", "Sopa de verduras", "Sushi variado"]
    }

    if 6 <= hora_local < 12:
        momento = "Desayuno"
    elif 12 <= hora_local < 16:
        momento = "Comida"
    elif 16 <= hora_local < 20:
        momento = "Merienda"
    else:
        momento = "Cena"

    print(f"\nEn {pais.capitalize()} son las {hora_local}:00")
    print(f"Es hora de: {momento}")
    print("Platos sugeridos:")
    for plato in opciones[momento]:
        print(f"- {plato}")

sugerir_comida_por_pais()
```
